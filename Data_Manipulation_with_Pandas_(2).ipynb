# For Data Selection by loc and iloc

```python
df.loc[:, ["Domestic Gross", "Production Budget"]]
```

```python
df.loc[1:4, ["Production Budget", "Rank"]]
```

```python
# loc --> we specified labels
# iloc --> we specified index number
df.loc[3, 'Rank']
```

```python
# first row then colum
df.iloc[4, [3,4]]
```

```python
df.rename(columns = {'Production Budget':'Production_Budget'} , inplace = True)
```

```python
df.loc[df.Production_Budget > 3000, : ]
```

```python
df.iloc[list(df.Production_Budget > 30000)]
```

```python
# With condition
df.loc[lambda df: df.Production_Budget > 5000,:]
```

```python
df.iloc[lambda df: [0,1], :3]
```

```python
df[df['Production_Budget'] > 0]
```

```python
df[df["Release Date"].isin(["12-Sep-12", "19-Jan-12"])]
```

```python
df["Release Date"].value_counts()
```
